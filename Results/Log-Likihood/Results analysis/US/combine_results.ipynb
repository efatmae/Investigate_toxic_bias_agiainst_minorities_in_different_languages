```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ALBERT_results_femm_sos = pd.read_csv("ALBERT/AlBERT_results_Femm_SOS.csv")
ALBERT_results_masc_sos = pd.read_csv("ALBERT/AlBERT_results_Masc_SOS.csv")
ALBERT_results_NB_sos = pd.read_csv("ALBERT/AlBERT_results_NB_SOS.csv")

BART_results_femm_sos = pd.read_csv("BART/BART_results_Femm_SOS.csv")
BART_results_masc_sos = pd.read_csv("BART/BART_results_Masc_SOS.csv")
BART_results_NB_sos = pd.read_csv("BART/BART_results_NB_SOS.csv")

BERT_results_femm_sos = pd.read_csv("BERT/BERT_results_Femm_SOS.csv")
BERT_results_masc_sos = pd.read_csv("BERT/BERT_results_Masc_SOS.csv")
BERT_results_NB_sos = pd.read_csv("BERT/BERT_results_NB_SOS.csv")

XLM_Roberta_results_femm_sos = pd.read_csv(
    "XLM_Roberta/XML_Roberta_results_Femm.csv")
XLM_Roberta_results_masc_sos = pd.read_csv(
    "XLM_Roberta/XML_Roberta_results_Masc.csv")
XLM_Roberta_results_NB_sos = pd.read_csv(
    "XLM_Roberta/XML_Roberta_results_NB.csv")
```

```python
combine_results = pd.concat([ALBERT_results_femm_sos,
                             ALBERT_results_masc_sos,
                             ALBERT_results_NB_sos,
                             
                             BART_results_femm_sos,
                             BART_results_masc_sos,
                             BART_results_NB_sos,
                             
                             BERT_results_femm_sos,
                             BERT_results_masc_sos,
                             BERT_results_NB_sos,
                             
                             XLM_Roberta_results_femm_sos,
                             XLM_Roberta_results_masc_sos,
                             XLM_Roberta_results_NB_sos
    
])
```

```python
combine_results["Marginalized_Bias_Score"] = combine_results["Marginalized_Bias_Score"].round(3)
combine_results["Non_Maarginalized_Bias_scores"] = combine_results["Non_Maarginalized_Bias_scores"].round(3)
```

```python
combine_results.to_csv("US_Combined_MLM_results.csv", index=False)
```

```python
combine_results["Mar_mins_Non_Mar"] = combine_results["Marginalized_Bias_Score"] - combine_results["Non_Maarginalized_Bias_scores"]
```

```python
combine_results["Mar_mins_Non_Mar"] = combine_results["Marginalized_Bias_Score"] - combine_results["Non_Maarginalized_Bias_scores"]
```

```python
SOS_high_aginst_marginalised.to_csv("SOS_high_aginst_marginalised.csv", index=False)
```

```python
SOS_high_aginst_non_marginalised = combine_results[combine_results["Marginalized_Bias_Score"] < combine_results["Non_Maarginalized_Bias_scores"]]
```

```python
SOS_high_aginst_non_marginalised.to_csv("SOS_high_aginst_non_marginalised.csv", index=False)
```

```python
Femm_data = combine_results[combine_results["Model_Name"].str.contains("_Femm")]
```

```python
Masc_data = combine_results[combine_results["Model_Name"].str.contains("_Masc")]
```

```python
NB_data = combine_results[combine_results["Model_Name"].str.contains("_NB")]
```

```python
fig, axs = plt.subplots(2,3, figsize=(15, 6))
fig.suptitle("SOS MLM Bias scores Englsih MLM (US)", fontweight="bold")

modelnames = ["ALBERT", "BART", "BERT", "XLM_Roberta_Multilingual"]

marginalized_femm = Femm_data.pivot(index='Model_Name', columns="Bias_Type", values="Marginalized_Bias_Score")
sns.heatmap(marginalized_femm, annot=True,  vmin=0.5, vmax=0.8, ax = axs[0,0], xticklabels=False, yticklabels=True)
axs[0,0].set_title("Femm-Marginalized", fontweight="bold")
axs[0,0].set_yticks(range(0,4), modelnames)
axs[0,0].set(xlabel="")
axs[0,0].set(ylabel="")

marginalized_masc = Masc_data.pivot(index='Model_Name', columns="Bias_Type", values="Marginalized_Bias_Score")
sns.heatmap(marginalized_masc, annot=True,  vmin=0.5, vmax=0.8, ax =axs[0,1], xticklabels=False, yticklabels=False)
axs[0,1].set(xlabel="")
axs[0,1].set_title("Masc-Marginalized", fontweight="bold")

marginalized_NB = NB_data.pivot(index='Model_Name', columns="Bias_Type", values="Marginalized_Bias_Score")
sns.heatmap(marginalized_NB, annot=True,  vmin=0.5, vmax=0.8, ax=axs[0,2], xticklabels=False, yticklabels=False)
axs[0,2].set(xlabel="")
axs[0,2].set_title("NB-Marginalized", fontweight="bold")

marginalized_femm = Femm_data.pivot(index='Model_Name', columns="Bias_Type", values="Non_Maarginalized_Bias_scores")
sns.heatmap(marginalized_femm, annot=True,  vmin=0.5, vmax=0.8, ax = axs[1,0])
axs[1,0].set_yticks(range(0,4), modelnames)
axs[1,0].set_title("Femm-Dominant", fontweight="bold")
axs[1,0].set(xlabel="")

marginalized_masc = Masc_data.pivot(index='Model_Name', columns="Bias_Type", values="Non_Maarginalized_Bias_scores")
sns.heatmap(marginalized_masc, annot=True,  vmin=0.5, vmax=0.8, ax =axs[1,1], xticklabels=True, yticklabels=False)
axs[1,1].set_title("Masc-Dominant", fontweight="bold")

marginalized_NB = NB_data.pivot(index='Model_Name', columns="Bias_Type", values="Non_Maarginalized_Bias_scores")
sns.heatmap(marginalized_NB, annot=True,  vmin=0.5, vmax=0.8, ax=axs[1,2], xticklabels=True, yticklabels=False)
axs[1,2].set(xlabel="")
axs[1,2].set_title("NB-Dominant", fontweight="bold")
```
