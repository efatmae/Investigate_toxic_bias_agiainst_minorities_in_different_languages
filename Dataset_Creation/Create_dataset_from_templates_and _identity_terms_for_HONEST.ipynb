```python
def create_temp_identity (sent_temp, identity_terms, bias_type_and_direction):
    sentences = []
    identity_terms = [x.strip() for x in identity_terms.split(",")]
    identity_terms = [x.lower() for x in identity_terms]
    identity = []
    for j in identity_terms:
        for i in sent_temp:
            if "ال[Identity]" in i:
                identity_article = "ال"+j
                sentences.append(i.replace("ال[Identity]", identity_article))
                print(identity_article)
                identity.append(j)
            else:
                sentences.append(i.replace("[Identity]", j))
                identity.append(j)
    bias_type = len(sentences) * [bias_type_and_direction]
    return sentences, bias_type, identity
```

```python
def create_sentence_temp_dataframe(temp, identity_terms, power_dynamics_flag=0):
    #Gender
    Mar_gender_sent, Mar_gender_bias_type_and_direction, Mar_gender_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Gender"]["Marginalized"].to_list()[0],
    "Gender_Marginalized")

    print("gender_Mar")
    print(len(Mar_gender_sent))

    print(len(Mar_gender_bias_type_and_direction))
    print(len(Mar_gender_identity))

    Non_Mar_gender_sent, Non_Mar_gender_bias_type_and_direction, Non_Mar_gender_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Gender"]["Non-Marginalized"].to_list()[0],"Gender_Non_marginalized")

    print("gender_NonM")
    print(len(Non_Mar_gender_sent))
    print(len(Non_Mar_gender_bias_type_and_direction))
    print(len(Non_Mar_gender_identity))

    #Race
    Mar_race_sent, Mar_race_bias_type_and_direction, Mar_race_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Race"]["Marginalized"].to_list()[0],"Race_Marginalized")

    print("race_Mar")
    print(len(Mar_race_sent))
    print(len(Mar_race_bias_type_and_direction))
    print(len(Mar_race_identity))
    
    Non_Mar_race_sent, Non_Mar_race_bias_type_and_direction, Non_Mar_race_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Race"]["Non-Marginalized"].to_list()[0],"Race_Non_marginalized")

    #Sexual Orientation
    Mar_SO_sent, Mar_SO_bias_type_and_direction, Mar_SO_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Sexual-Orientation"]["Marginalized"].to_list()[0],"Sexual_orientation_Marginalized")
    
    print("SO_Mar")
    print(len(Mar_SO_sent))
    print(len(Mar_SO_bias_type_and_direction))
    print(len(Mar_SO_identity))
    
    Non_Mar_SO_sent, non_Mar_SO_bias_type_and_direction, Non_Mar_SO_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Sexual-Orientation"]["Non-Marginalized"].to_list()[0],"Sexual_orientation_Non_Marginalized")

    #Disability
    Mar_disability_sent, Mar_disability_bias_type_and_direction, Mar_disability_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Disability"]["Marginalized"].to_list()[0],"Physical_ability_Marginalized")

    print("DIS_Mar")
    print(len(Mar_disability_sent))
    print(len(Mar_disability_bias_type_and_direction))
    print(len(Mar_disability_identity))
    
    Non_Mar_disability_sent, non_Mar_disability_bias_type_and_direction, Non_Mar_disability_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Disability"]["Non-Marginalized"].to_list()[0],"Physical_ability_Non_Marginalized")

    #Social Class
    #Mar_SC_toxic_sent, Mar_SC_non_toxic_sent, Mar_SC_bias_type_and_direction, Mar_SC_identity = create_temp_identity(temp["toxic"], temp["non-toxic"], identity_terms[identity_terms["attribute"]=="Income"]["Marginalized"].to_list()[0],"Social-Class_Marginalized")

    #print("Income_Mar")
    #print(len(Mar_SC_toxic_sent))
    #print(len(Mar_SC_non_toxic_sent))
    #print(len(Mar_SC_bias_type_and_direction))
    #print(len(Mar_SC_identity))
    
    #Non_Mar_SC_toxic_sent, Non_Mar_SC_non_toxic_sent, non_Mar_SC_bias_type_and_direction, Non_Mar_SC_identity = create_temp_identity(temp["toxic"], temp["non-toxic"], identity_terms[identity_terms["attribute"]=="Income"]["Non-Marginalized"].to_list()[0],"Social-Class_Non_Marginalized")

    #Religion
    Mar_religion_sent, Mar_religion_bias_type_and_direction, Mar_religion_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Religion"]["Marginalized"].to_list()[0],"Religion_Marginalized")

    print("religion_Mar")
    print(len(Mar_religion_sent))
    print(len(Mar_religion_bias_type_and_direction))
    print(len(Mar_religion_identity))
    
    Non_Mar_religion_sent, non_Mar_religion_bias_type_and_direction, Non_Mar_religion_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Religion"]["Non-Marginalized"].to_list()[0],"Religion_Non_Marginalized")
    
    print("religion_NON_Mar")
    print(len(Non_Mar_religion_sent))
    print(len(non_Mar_religion_bias_type_and_direction))
    print(len(Non_Mar_religion_identity))
    
    if power_dynamics_flag==1:
     #Refugees
        Mar_refugees_sent, Mar_refugees_bias_type_and_direction, Mar_refugees_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Refugees"]["Marginalized"].to_list()[0],"Refugees_Marginalized")
    
        print("refugges_Mar")
        print(len(Mar_refugees_sent))
        print(len(Mar_refugees_bias_type_and_direction))
        print(len(Mar_refugees_identity))
        
        Non_Mar_refugees_sent,non_Mar_refugees_bias_type_and_direction, Non_Mar_refugees_identity = create_temp_identity(temp["sentence"], identity_terms[identity_terms["attribute"]=="Refugees"]["Non-Marginalized"].to_list()[0],"Refugees_Non_Marginalized")

    data_SO_Mar = pd.DataFrame({"sent": Mar_SO_sent,
                                "bias_type": Mar_SO_bias_type_and_direction,
                                   "identity": Mar_SO_identity,
                                "stereo_antistereo": "stereo"})

    data_SO_Non_Mar = pd.DataFrame({"sent": Non_Mar_SO_sent,
                                "bias_type": non_Mar_SO_bias_type_and_direction,
                                   "identity": Non_Mar_SO_identity,
                                "stereo_antistereo": "stereo" })

    data_gender_Mar = pd.DataFrame({"sent": Mar_gender_sent,
                                "bias_type": Mar_gender_bias_type_and_direction,
                             "identity": Mar_gender_identity,
                                "stereo_antistereo": "stereo" })

    data_gender_Non_Mar = pd.DataFrame({"sent": Non_Mar_gender_sent,
                                "bias_type": Non_Mar_gender_bias_type_and_direction,
                             "identity": Non_Mar_gender_identity,
                                "stereo_antistereo": "stereo" })

    data_race_Mar = pd.DataFrame({"sent": Mar_race_sent,
                                "bias_type": Mar_race_bias_type_and_direction,
                             "identity": Mar_race_identity,
                                "stereo_antistereo": "stereo" })

    data_race_Non_Mar = pd.DataFrame({"sent": Non_Mar_race_sent,
                                "bias_type": Non_Mar_race_bias_type_and_direction,
                             "identity": Non_Mar_race_identity,
                                "stereo_antistereo": "stereo" })


    data_religion_Mar = pd.DataFrame({"sent": Mar_religion_sent,
                                "bias_type": Mar_religion_bias_type_and_direction,
                             "identity": Mar_religion_identity,
                                "stereo_antistereo": "stereo" })

    data_religion_Non_Mar = pd.DataFrame({"sent": Non_Mar_religion_sent,
                                "bias_type": non_Mar_religion_bias_type_and_direction,
                             "identity": Non_Mar_religion_identity,
                                "stereo_antistereo": "stereo" })

    #data_social_class_Mar = pd.DataFrame({"Toxic_sent": Mar_SC_toxic_sent,
                              # "Non_Toxic_sent": Mar_SC_non_toxic_sent,
                               # "bias_type": Mar_SC_bias_type_and_direction,
                             # "identity": Mar_SC_identity,
                               # "stereo_antistereo": "stereo" })

    #data_social_class_Non_Mar = pd.DataFrame({"Toxic_sent": Non_Mar_SC_toxic_sent,
     #                          "Non_Toxic_sent": Non_Mar_SC_non_toxic_sent,
      #                          "bias_type": non_Mar_SC_bias_type_and_direction,
       #                      "identity": Non_Mar_SC_identity,
        #                        "stereo_antistereo": "stereo" })

    data_disability_Mar = pd.DataFrame({"sent": Mar_disability_sent,
                                "bias_type": Mar_disability_bias_type_and_direction,
                             "identity": Mar_disability_identity,
                                "stereo_antistereo": "stereo" })

    data_disability_Non_Mar = pd.DataFrame({"sent": Non_Mar_disability_sent,
                                "bias_type": non_Mar_disability_bias_type_and_direction,
                             "identity": Non_Mar_disability_identity,
                                "stereo_antistereo": "stereo" })

    if power_dynamics_flag==1:
            data_refugees_Mar = pd.DataFrame({"sent": Mar_refugees_sent,
                                "bias_type": Mar_refugees_bias_type_and_direction,
                             "identity": Mar_refugees_identity,
                                "stereo_antistereo": "stereo" })
            
            data_refugees_Non_Mar = pd.DataFrame({"sent": Non_Mar_refugees_sent,
                                "bias_type": non_Mar_refugees_bias_type_and_direction,
                             "identity": Non_Mar_refugees_identity,
                                "stereo_antistereo": "stereo" })
    
    if power_dynamics_flag==1:
        SOS_data = pd.concat([data_SO_Mar, data_SO_Non_Mar,
                         data_gender_Mar, data_gender_Non_Mar,
                         data_race_Mar, data_race_Non_Mar,
                         data_religion_Mar, data_religion_Non_Mar,
                         #data_social_class_Mar, data_social_class_Non_Mar,
                         data_disability_Mar, data_disability_Non_Mar,
                         data_refugees_Mar, data_refugees_Non_Mar])
        
    else:
        SOS_data = pd.concat([data_SO_Mar, data_SO_Non_Mar,
                         data_gender_Mar, data_gender_Non_Mar,
                         data_race_Mar, data_race_Non_Mar,
                         data_religion_Mar, data_religion_Non_Mar,
                        # data_social_class_Mar, data_social_class_Non_Mar,
                         data_disability_Mar, data_disability_Non_Mar])
        
    
    SOS_data = SOS_data.sample(frac=1)
    print(len(SOS_data))
    return SOS_data
```
